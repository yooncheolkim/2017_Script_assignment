[2번 문제]

다형성이란 하나의 클래스에서 상속을 이용하여, 여러개의 자료형 타입, 즉 기존 클래스가 다양하게 변화된 자식 클래스들을 만들 수 있는 성질을 다형성이라고 한다. 예를 들어 bird 라는 클래스가 있으면, 이 bird 클래스의 상속되어진 여러 형태의 닭, 까마귀, 독수리 등의 클래스를 생성할 수 있는 성질이다.

```python
class Bird:
    isflying = True
    crying = 'bbak bbak'
class Chicken(Bird):
    def __init__(self, flyable = False, crying = 'koko'):
        self.isflying = flyable
        self.crying = crying
class Eagle(Bird):
    def __init__(self, flyable = True, crying = 'kkio'):
        self.isflying = flyable
        self.crying = crying
b = Bird()
c = Chicken()
e = Eagle()
print(b.crying)
print(c.crying)
print(e.crying)
```

[3번 문제]

```python
class Counter:
    def __init__(self,count,step = 1):
        self.count = count
        self.step = step
    def __str__(self):
        return "[COUNT(step: " + str(self.step) + ")] " + str(self.count)
    def incr(self):
        self.count += self.step
    def __call__(self):
        self.incr()
    def __add__(self,other):
        self.count += other
        return self
    def __sub__(self,other):
        self.count -= other
        return self
    def __lt__(self,other):
        return self.count < other
    def __gt__(self,other):
        return self.count > other
    def __eq__(self,other):
        return self.count == other
    def __repr__(self):
        return self.count
```

```python
c = Counter(10)
d = Counter(10,2)
```

```python
print(c)
print(d)
```

```python
c.incr()
d.incr()
print(c)
print(d)
```

```python
c()
d()
print(c)
print(d)
```

```python
c = c + 5
d = d - 5
print(c)
print(d)
```

```python
print(c > 10)
print(d > 10) 
print(c < 10 )
```

- \_\_init\_\_ 메소드는 클래스가 생성될때 호출되는 생성자 메소드 이다. 
 즉, s = MySet([1,2,2,3)을 할때 __init__메소드가 호출된다. 메소드 안을 보면, list를 상속받은 Myset은 매개변수의 것들을 집어넣고, eliminate_duplicate를 실행한다.
 
- eliminate_duplicate 메소드는 말 그대로, 중복되는 값을 삭제 한다. 현재 가지고 있는 list에서 빈 list인 s에 하나씩 옮겨 넣는데, 이때 s에 중복된 값이 있으면 넣지 않는다. 이후, self를 비운후, self에 다시 s를 옮겨 넣는다. 이 메소드는 staticmethod 이므로, 인스턴스와 무관하게 클래스 이름 공간에 존재하는 메소드이다. 그러므로, 인스턴스 뿐만 아니라, 클래스 이름을 사용하여서도 호출 가능하다.

- \_\_str\_\_ 메소드는 현재 Myset 이 가지고 있는 원소를 보여주는 메소드 이다. 
 result 에 틀을 만들고,  for문을 이용하여 원소 하나씩 str + 연산을 이용하여 추가한다. 그리고 마지막 ' ,'를 제외하고 '}'를 추가하여 result를 반환한다.
 

[5번 문제]

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    def __or__(self,other):
        result = MySet([])
        # result에 self 넣기
        for e in self:
            result.append(e)
        # result에 other 넣기
        for e in other:
            result.append(e)
        # 중복되는 값 삭제
        MySet.eliminate_duplicate(result)
        return result
    
    def __and__(self,other):
        result = MySet([])
        # self에 other의 값이 있으면 result에 넣기
        for e in other:
            if e in self:
                result.append(e)
        return result
    def __sub__(self,other):
        # result 에 self 값 복사
        result = MySet([])
        for e in self:
            result.append(e)
        # result에 other의 값이 있으면, result에서 그 값 삭제
        for e in other:
            if e in result:
                result.remove(e)
        return result
                
s = MySet([1, 2, 2, 3])
print(s)
t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
print(t)
```

```python
from urllib import request
import string

class WebWordsFrequency:
    def __init__(self,*url):
        self.my_url = []
        for i in url:
            self.my_url.append(i)
            
    def addUrl(self,url):
        if url not in self.my_url:
            self.my_url.append(url)
    
    def removeUrl(self,url):
        if url in self.my_url:
            self.my_url.remove(url)
        
    def listUrls(self):
        for i in self.my_url:
            print(i)
    
    def getWordsFrequency(self):
        self.url_frequency = {}
        self.such_url_frequency = []
        # 단어들을 임시로 temp 에 저장
        temp = []
        ## url에 있는 source에서 사용할수 있는 단어 뽑어 내서 temp에 저장
        for k in self.my_url:
            temp.append(self.extractionFromSource(k))
        
        #temp 에서 구두 문자 제거
        for k in range(0,len(temp)): # 이거 조심...! k는 인덱스 인데 리스트로 받으면 안됨..
            for j in range(0,len(temp[k])):
                temp[k][j] = self.punctuationToEmpty(temp[k][j])
                
        # 불용어 제거
        for k in range(0,len(temp)):
            temp[k] = self.stopwordToEmpty(temp[k])
            
        # temp에서 빈 문자열 제거
        for k in range(0,len(temp)):
            temp[k] = self.deleteEmpty(temp[k])
            
        # 각 url의 빈도수를 따로따로 측정, such_url_frequency 에 저장
        for k in range(0,len(temp)):
            self.such_url_frequency.append(self.such_frequency(temp[k]))
        
        # temp에서 각 요소값의 빈도수를 사전으로 반환하여, url_frequency에 추가
        for k in range(0,len(temp)):
            self.all_frequency(temp[k])
            
        return self.url_frequency
            
    def extractionFromSource(self,url):
        f = request.urlopen(url)
        source_bytes = f.read()
        source = source_bytes.decode('utf-8')
        
        isOpen = False
        pureWord = []
        text = ''
        for i in source:
            if i == '<':
                isOpen = True
                continue
            if i == '>':
                isOpen = False
                continue
            if isOpen == False and not i.isspace():
                text = text + i
            if isOpen == False and i.isspace() and text:
                pureWord.append(text)
                text=''
        return pureWord
    
    def stopwordToEmpty(self,a):
        # 불용어 추출
        url1 = 'http://www.lextek.com/manuals/onix/stopwords1.html'
        f = request.urlopen(url1)
        source_bytes = f.read()
        source = source_bytes.decode('utf-8')
        L = source.split()
        L = L[503:503+429]
        
        # 불용어, a에서 empty 로 만듬.
        for i in L:
            for j in range(0,len(a)):
                temp = a[j].lower()
                if i == temp:
                    a[j] = ''
        return a
        
    def punctuationToEmpty(self,a):
        # 구두문자 제거
        L = [string.punctuation[i] for i in range(0,len(string.punctuation)-1)]
        list_a = list(a)
        for i in range(0,len(list_a)):
            for j in range(0,len(L)):
                if list_a[i] == L[j]:
                    list_a[i] = ''
        result = ""
        for i in range(0,len(list_a)):
            result = result + list_a[i]
        return result
    
    def deleteEmpty(self,a):
        while( '' in a):
            a.remove('')
        return a
    
    #종합한 빈도
    def all_frequency(self,a):
        for i in range(0,len(a)):
            if (self.url_frequency.get(a[i])):
                self.url_frequency[a[i]] += 1 ## 기존에 있는 key 면 value 값 1 증가
            else:
                self.url_frequency[a[i]] = 1 ##사전에 a[i] 와 value 1추가
                
    # 따로 따로 빈도, 사전을 반환
    def such_frequency(self,a):
        result = {}
        for i in range(0,len(a)):
            if (result.get(a[i])):
                result[a[i]] += 1
            else:
                result[a[i]] = 1
        return result
    # url_frequency 에서 가장 많은 단어의 빈도수를 보이는 값 반환
    def getMaxFrequencyWords(self):
        if self.my_url == []:
            return None
        max_frequency = 0
        result = []
        for i in self.url_frequency.values():
            max_frequency = i if max_frequency < i else max_frequency # 삼항 연산자
        for i in self.url_frequency.keys():
            if self.url_frequency[i] == max_frequency:
                result.append(i)
        return result
    
    def searchUrlByWord(self,a):
        similarity =[0 for i in range(0,len(self.my_url))]
        url = []
        for i in range(0,len(self.such_url_frequency)):
            if a in self.such_url_frequency[i].keys():
                similarity[i] = self.such_url_frequency[i][a]
        
        max_frequency = similarity[0]
        for i in similarity:
            max_frequency = i if max_frequency < i else max_frequency
        for i in range(0,len(similarity)):
            if max_frequency == similarity[i]:
                url.append(self.my_url[i])
        return url
```
